```python
import pandas as pd
from enum import Enum

class Attr: # attributes enum for clarity
    SEPALS = 0
    PETAL = 2
    KIND = 4
    
df = pd.read_csv("iris.data", header=None)
df.tail() 
# data check, outputs last 5 rows
# TODO: It doesn't output row with index 149 !!!
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = df.iloc[0:100, Attr.KIND].values # take KIND column from dataset
y = np.where(y == 'Iris-setosa', -1, 1) # split dataset into two by kind of iris 

x = df.iloc[0:100, [Attr.SEPALS, Attr.PETAL]].values # take SEPALS and PETAL length columns

plt.scatter(
    x[0:50, 0], x[0:50, 1],
    color = "red", marker=".", label="setosa",
)

plt.scatter(
    x[50:100, 0], x[50:100, 1],
    color = "blue", marker=".", label="versicolor",
)
plt.xlabel("sepals length")
plt.ylabel("petal length")
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from perceptron import Perceptron

# import perceptron
# from importlib import reload
# reload(perceptron)

y = df.iloc[0:100, Attr.KIND].values
y = np.where(y == 'Iris-setosa', -1, 1)
x = df.iloc[0:100, [Attr.SEPALS, Attr.PETAL]].values

ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(x, y)

plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_, marker="o")
plt.xlabel("Ages")
plt.ylabel("False classifications count")
plt.show()
```
